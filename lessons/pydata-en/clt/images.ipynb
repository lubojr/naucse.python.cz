### Obrázky k teoretické části

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from random import randrange
%matplotlib inline
```

```python
seznam = []
for x in range(1, 7):
    seznam.extend([x]*10)
```

```python
d = pd.DataFrame(seznam, columns=["výsledek"])
```

```python
d.groupby("výsledek").size().plot.bar(figsize=(10,5));
plt.title("Rozdělení hodů kostkou");
plt.xlabel("Hod");
plt.ylabel("Počet výskytů");
plt.savefig("static/uniform.png")
```

```python
hmotnost = np.random.normal(74, 14.5, size=100000).tolist()
```

```python
dh = pd.DataFrame(hmotnost, columns=["hmotnost"])
```

```python
dh.hist(bins=100, figsize=(10,5), grid=False)
plt.title("Hmotnost - normální rozložení")
plt.xlabel("Hmotnost")
plt.ylabel("Počet výskytů")
plt.savefig("static/normal.png")
```

```python
porucha = np.random.exponential(size=100000).tolist()
```

```python
dp = pd.DataFrame(porucha, columns=["porucha"])
```

```python
dp.hist(bins=50, figsize=(10,5), grid=False)
plt.title("Porucha zařízení - exponencíální rozložení")
plt.xlabel("Počet let")
plt.ylabel("Počet zařízení bez vady")
plt.savefig("static/exponential.png")
```
